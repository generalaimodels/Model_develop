````python
from IPython.display import Image, display
import os

# Set the desired height and width
height = 300
width = 500
base_folder = "/content/machine-learning-book"
chapters = [f"ch{i:02d}" for i in range(1, 20)]

for chapter in chapters:
    folder_path = os.path.join(base_folder, chapter, "figures")

    # List all files in the folder
    files = os.listdir(folder_path)

    # Filter only image files (you can customize the extensions if needed)
    image_files = [file for file in files if file.lower().endswith(('.png', '.jpg', '.jpeg'))]

    # Sort image files alphabetically
    image_files.sort()

    # Display each image for the current chapter
    print(f"Chapter: {chapter}")
    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        display(Image(filename=image_path, height=height, width=width))
````

````python
from IPython.display import Image
import os
from PIL import Image as PILImage  # PIL library for image processing

# Set the desired height and width
height = 300
width = 500

# Specify the output folder
output_folder = "/content/output_images"

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

base_folder = "/content/machine-learning-book"
chapters = [f"ch{i:02d}" for i in range(1, 20)]

for chapter in chapters:
    folder_path = os.path.join(base_folder, chapter, "figures")

    # List all files in the folder
    files = os.listdir(folder_path)

    # Filter only image files (you can customize the extensions if needed)
    image_files = [file for file in files if file.lower().endswith(('.png', '.jpg', '.jpeg'))]

    # Sort image files alphabetically
    image_files.sort()

    # Display and save each image for the current chapter
    print(f"Saving images for Chapter: {chapter}")
    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)

        # Create a unique filename for the saved image
        output_filename = os.path.join(output_folder, f"{chapter}_{image_file}")

        # Display and save the image
        display(Image(filename=image_path, height=height, width=width))
        pil_image = PILImage.open(image_path)
        pil_image.save(output_filename)

print("Images saved to:", output_folder)
````

The `jupyter nbconvert` command allows you to convert Jupyter Notebooks from one format to another. You can specify the desired output format using the `--to` option. Here are some common output formats:

1. **HTML:**
   ```bash
   jupyter nbconvert --to html your_notebook.ipynb
   ```

2. **PDF:**
   ```bash
   jupyter nbconvert --to pdf your_notebook.ipynb
   ```

3. **Python Script:**
   ```bash
   jupyter nbconvert --to script your_notebook.ipynb
   ```

4. **Markdown:**
   ```bash
   jupyter nbconvert --to markdown your_notebook.ipynb
   ```

5. **Reveal.js Slideshow (HTML slides):**
   ```bash
   jupyter nbconvert --to slides your_notebook.ipynb --post serve
   ```

6. **Latex:**
   ```bash
   jupyter nbconvert --to latex your_notebook.ipynb
   ```

These are just a few examples, and there are more output formats available. You can see the full list of supported formats by running:

```bash
jupyter nbconvert --help
```

This command will display information about various options, including the available output formats. You can also specify additional options for customization, such as template files, CSS styles, and more, depending on the output format you choose.
