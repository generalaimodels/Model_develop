```python
def count_unique_words(string):
    # Remove punctuation and make the string lowercase
    string = string.lower().translate(str.maketrans("", "", string.punctuation))
    # Split the string into individual words
    words = string.split()
    # Create a dictionary to store the word counts
    word_counts = {}
    # Iterate over the words and increment the count for each unique word
    for word in words:
        if word not in word_counts:
            word_counts[word] = 0
        word_counts[word] += 1
    # Return the word counts, sorted by frequency in descending order
    return {word: count for word, count in word_counts.items() if count > 0}
```

```python
print(count_unique_words('This is a sample sentence. This sentence is just an example.'))
```

```python
import random

def is_prime(n):
    if n < 2:
        return False
    for _ in range(5):
        a = random.randint(2, n - 2)
        x = pow(a, n - 1, n)
        if x == 1 or x == n - 1:
            continue
        for _ in range(5):
            x = pow(x, 2, n)
            if x == n - 1:
                break
        else:
            return False
    return True
```

```python
for n in range(10 ** 19, 10 ** 20):
    if is_prime(n):
        print(n)
        break
else:
    print("No prime found")
```

```python
from bs4 import BeautifulSoup
import requests

# URL of the website to search
url = "https://www.python.org/"

# Keyword to search for
keyword = "python"

# Send a request to the website and get the HTML response
response = requests.get(url)
html = response.text

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Find all the elements that contain the keyword
results = soup.find_all('div', {'class': 'result'})

# Iterate over the results and extract the relevant information
for result in results:
    title = result.find('h3').text.strip()
    link = result.find('a')['href']
    description = result.find('p').text.strip()
    print(title, link, description)
```
