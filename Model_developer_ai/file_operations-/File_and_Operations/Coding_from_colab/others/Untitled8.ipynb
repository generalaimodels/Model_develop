```python
from typing import Union,List, Dict, Optional
from datasets import load_dataset, DatasetDict
from transformers import AutoTokenizer
import multiprocessing
import warnings
def advanced_data_loader(input: Union[str, Dict[str, str]], format: Optional[str] = None, split_ratios: Optional[Dict[str, float]] = None) -> Optional[DatasetDict]:
    """
    Loads a dataset from a given input path or dictionary specifying file paths and splits it.

    :param input: A string representing the dataset name or directory, or a dictionary containing file paths.
    :param format: The format of the dataset if loading from a file (e.g., 'csv' or 'json').
    :param split_ratios: A dictionary with keys 'train', 'test', and 'eval' containing split ratios.
    :return: A loaded and split dataset or None in case of failure.
    """
    if split_ratios is None:
        split_ratios = {'train': 0.8, 'test': 0.1, 'eval': 0.1}

    try:
        # Load the dataset
        if isinstance(input, dict) and format in ['csv', 'json']:
            dataset = load_dataset(format, data_files=input)
        elif isinstance(input, str) and format == 'text':
            dataset = load_dataset(format, data_dir=input)
        elif isinstance(input, str) and format is None:
            dataset = load_dataset(input)
        else:
            warnings.warn("Invalid input or format. Please provide a valid dataset name, directory, or file paths.")
            return None
    except FileNotFoundError as e:
        warnings.warn(str(e))
        return None

    # Split the dataset
    if dataset:
        split_dataset = dataset['train'].train_test_split(test_size=split_ratios['test'] + split_ratios['eval'])
        test_eval_dataset = split_dataset['test'].train_test_split(test_size=split_ratios['eval'] / (split_ratios['test'] + split_ratios['eval']))
        dataset = DatasetDict({
            'train': split_dataset['train'],
            'test': test_eval_dataset['train'],
            'eval': test_eval_dataset['test']
        })

    print("Splits: ", dataset.keys())
    print("Columns: ", {split: dataset[split].column_names for split in dataset.keys()})
    return dataset
```

```python
%%time
from typing import List, Any, Dict, Tuple, Optional

def extract_text_samples(dataset: Dict[str, Any], batch_size: int, columns:List[str]) -> Tuple[List[str], List[Any]]:
    """
    Function to extract sample text data from columns in a dataset.

    Args:
        dataset: A dictionary where keys are column names and values are lists of column data.
        batch_size: The size of the batch for the output.

    Returns:
        A tuple containing a list of column names with text-like content and a list of corresponding sample text data.
    """
    text_columns = []
    text_samples = []

    # Iterate through column names and collect column names with text-like samples and their data
    for column in columns:
        first_non_null = next((val for val in dataset['train'][column] if val is not None), None)
        if isinstance(first_non_null, str):
            # Column with string type
            text_columns.append(column)
            text_samples.append(dataset['train'][column][:batch_size])
        elif isinstance(first_non_null, list) and all(isinstance(item, str) for item in first_non_null):
            # Column with list of strings
            text_columns.append(column)
            text_samples.append(dataset['train'][column][:batch_size])
        elif isinstance(first_non_null, dict) and all(isinstance(val, str) for val in first_non_null.values()):
            # Column with dictionary of strings
            text_columns.append(column)
            text_samples.append([val for _, val in dataset['train'][column].items()][:batch_size])
        elif isinstance(first_non_null, list) and all(isinstance(item, dict) for item in first_non_null):
            # Column with list of dictionaries
            text_columns.append(column)
            text_samples.append([[ d] for d in dataset['train'][column][:batch_size]])
        elif isinstance(first_non_null, dict) and all(isinstance(val, list) for val in first_non_null.values()):
            # Column with dictionary of lists of strings
            text_columns.append(column)
            text_samples.append([val for _, val in dataset['train'][column].items()][:batch_size])

    return text_columns, text_samples
batch_size_example = 10
text_columns, text_content = extract_text_samples(dataset, batch_size_example, columns= ['act', 'prompt'])
print(text_columns, text_content)

```

```python
from typing import List
from transformers import AutoTokenizer
from tqdm import tqdm

# Replace 'your-model-name-here' with the actual model name you're using.
MODEL = 'gpt2'
tokenizer = AutoTokenizer.from_pretrained(MODEL, trust_remote_code=True)

def chars_token_ratio(texts: List[str], tokenizer, nb_examples: int = 400) -> float:
    """
    Estimate the average number of characters per token for a list of strings.

    Args:
        texts (List[str]): List of text strings to be analyzed.
        tokenizer: Tokenizer object used for tokenizing the text.
        nb_examples (int, optional): The number of examples to consider for the estimation.

    Returns:
        float: The average number of characters per token.

    Raises:
        ValueError: If the input is not a list of strings or is empty.

    Usage example:
        sample_texts = ["This is a sample sentence.", "Here's another one."]
        chars_per_token = chars_token_ratio(sample_texts, tokenizer, nb_examples=2)
        print(f"The character to token ratio is: {chars_per_token:.2f}")
    """
    if not all(isinstance(text, str) for text in texts):
        raise ValueError("All items in the input list must be strings.")
    if not texts:
        raise ValueError("The input list cannot be empty.")

    total_characters, total_tokens = 0, 0
    for text in tqdm(texts[:min(nb_examples, len(texts))], total=min(nb_examples, len(texts))):
        total_characters += len(text)
        total_tokens += len(tokenizer.tokenize(text))

    if total_tokens == 0:
        raise ValueError("The total number of tokens is zero, cannot divide by zero.")

    return total_characters / total_tokens

# Sample usage:
sample_texts = text_content[1]
try:
    chars_per_token = chars_token_ratio(sample_texts, tokenizer, nb_examples=2)
    print(f"The character to token ratio is: {chars_per_token:.2f}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import functools
import numpy as np
from transformers import PreTrainedTokenizer

# Helper function to get token ids of the special tokens for prefix, suffix, and middle for FIM transformations.
@functools.lru_cache(maxsize=None)
def get_fim_token_ids(tokenizer: PreTrainedTokenizer):
    """
    Get the token IDs for the FIM special tokens.

    Args:
        tokenizer (PreTrainedTokenizer): The tokenizer with the fim special tokens.

    Returns:
        tuple: Token IDs for FIM_SUFFIX, FIM_PREFIX, FIM_MIDDLE, FIM_PAD.
    """
    try:
        FIM_PREFIX, FIM_MIDDLE, FIM_SUFFIX, FIM_PAD = tokenizer.special_tokens_map["additional_special_tokens"][1:5]
        suffix_tok_id, prefix_tok_id, middle_tok_id, pad_tok_id = (
            tokenizer.vocab[tok] for tok in [FIM_SUFFIX, FIM_PREFIX, FIM_MIDDLE, FIM_PAD]
        )
    except KeyError:
        suffix_tok_id, prefix_tok_id, middle_tok_id, pad_tok_id = None, None, None, None
    return suffix_tok_id, prefix_tok_id, middle_tok_id, pad_tok_id


def permute(
    sample: list,
    np_rng: np.random.Generator,
    suffix_tok_id: int,
    prefix_tok_id: int,
    middle_tok_id: int,
    pad_tok_id: int,
    fim_rate: float = 0.5,
    fim_spm_rate: float = 0.5,
    truncate_or_pad: bool = False,
) -> (list, np.random.Generator):
    """
    Perform a FIM transformation on a list of tokens.

    Args:
        sample (list): List of token IDs to be transformed.
        np_rng (np.random.Generator): NumPy random generator for reproducibility.
        suffix_tok_id (int): Token ID for FIM_SUFFIX.
        prefix_tok_id (int): Token ID for FIM_PREFIX.
        middle_tok_id (int): Token ID for FIM_MIDDLE.
        pad_tok_id (int): Token ID for FIM_PAD.
        fim_rate (float): Probability of applying FIM transformation.
        fim_spm_rate (float): Probability of applying SPM variant of FIM transformation.
        truncate_or_pad (bool): Whether to truncate or pad the transformed sequence to match original length.

    Returns:
        tuple: The transformed sample and the NumPy random generator.
    """
    # Apply FIM transformation with a probability of fim_rate
    if np_rng.binomial(1, fim_rate):
        # Determine random split points for the sample
        boundaries = sorted(np_rng.choice(len(sample) + 1, size=2, replace=False))
        prefix, middle, suffix = sample[:boundaries[0]], sample[boundaries[0]:boundaries[1]], sample[boundaries[1]:]

        if truncate_or_pad:
            # Calculate the new length and adjust suffix length if needed
            new_length = len(prefix) + len(middle) + len(suffix) + 3  # 3 for special tokens
            diff = new_length - len(sample)
            suffix = suffix[:len(suffix) - diff] if diff > 0 else np.pad(suffix, (0, -diff), constant_values=pad_tok_id)

        # Apply SPM variant with probability fim_spm_rate, otherwise apply PSM variant
        if np_rng.binomial(1, fim_spm_rate):
            new_sample = [prefix_tok_id, suffix_tok_id] + suffix + [middle_tok_id] + prefix + middle
        else:
            new_sample = [prefix_tok_id] + prefix + [suffix_tok_id] + suffix + [middle_tok_id] + middle

        return new_sample, np_rng

    # Return the original sample if no transformation is applied
    return sample, np_rng
```

```python
from typing import List, Any, Dict, Tuple
from multiprocessing import Pool, cpu_count

def extract_column_data(column_data, batch_size):
    first_non_null = None
    for val in column_data:
        if val is not None:
            first_non_null = val
            break

    if isinstance(first_non_null, str):
        return column_data[:batch_size]
    elif isinstance(first_non_null, list) and all(isinstance(item, str) for item in first_non_null):
        return column_data[:batch_size]
    elif isinstance(first_non_null, dict) and all(isinstance(val, str) for val in first_non_null.values()):
        return list(first_non_null.values())[:batch_size]
    elif isinstance(first_non_null, list) and all(isinstance(item, dict) for item in first_non_null):
        sample_data = []
        for item in column_data[:batch_size]:
            sample_data.extend([d] for d in item if isinstance(d, dict))
        return sample_data
    elif isinstance(first_non_null, dict) and all(isinstance(val, list) for val in first_non_null.values()):
        sample_data = []
        for val in first_non_null.values():
            if isinstance(val, list) and all(isinstance(sub_val, str) for sub_val in val):
                sample_data.extend(val)
        return sample_data[:batch_size]
    return []

def extract_text_samples_parallel(dataset: Dict[str, Any], batch_size: int) -> Tuple[List[str], List[Any]]:
    text_columns = []
    text_samples = []

    train_data = dataset['train']

    # Define a pool of workers equal to the number of available CPU cores
    with Pool(cpu_count()) as pool:
        # Prepare the data for the worker function
        jobs = [(train_data[column], batch_size) for column in train_data.column_names]

        # Map the extract_column_data function onto the data in parallel
        results = pool.starmap(extract_column_data, jobs)

    for i, column in enumerate(train_data.column_names):
        if results[i]:
            text_columns.append(column)
            text_samples.append(results[i])

    return text_columns, text_samples

# Example Usage
if __name__ == "__main__":
    batch_size_example = 10
    # dataset = ...  # Initialize your dataset here
    text_columns, text_content = extract_text_samples_parallel(dataset, batch_size_example)
    print(text_columns, text_content)
```

```python
# Import the required modules

from typing import List, Any, Dict, Tuple, Optional
import multiprocessing as mp

def extract_text_samples(dataset: Dict[str, Any], batch_size: int) -> Tuple[List[str], List[Any]]:
    """
    Function to extract sample text data from columns in a dataset.

    Args:
        dataset: A dictionary where keys are column names and values are lists of column data.
        batch_size: The size of the batch for the output.

    Returns:
        A tuple containing a list of column names with text-like content and a list of corresponding sample text data.
    """

    # Initialize variables to store text columns and samples
    text_columns = []
    text_samples = []

    # Define a worker function to process each column in parallel
    def worker(column_data):
        column, data = column_data
        first_non_null = next((val for val in data if val is not None), None)
        if isinstance(first_non_null, str):
            # Column with string type
            return column, data[:batch_size]
        elif isinstance(first_non_null, list) and all(isinstance(item, str) for item in first_non_null):
            # Column with list of strings
            return column, data[:batch_size]
        elif isinstance(first_non_null, dict) and all(isinstance(val, str) for val in first_non_null.values()):
            # Column with dictionary of strings
            return column, [val for _, val in data.items()][:batch_size]
        elif isinstance(first_non_null, list) and all(isinstance(item, dict) for item in first_non_null):
            # Column with list of dictionaries
            return column, [[d] for d in data[:batch_size]]
        elif isinstance(first_non_null, dict) and all(isinstance(val, list) for val in first_non_null.values()):
            # Column with dictionary of lists of strings
            return column, [val for _, val in data.items()][:batch_size]
        return None

    # Create a multiprocessing pool
    pool = mp.Pool()

    # Iterate through column names and collect column names with text-like samples and their data
    for result in pool.imap_unordered(worker, dataset['train'].items()):
        if result:
            text_columns.append(result[0])
            text_samples.append(result[1])

    return text_columns, text_samples

# Example Usage
if __name__ == "__main__":
    batch_size_example = 10
    text_columns, text_content = extract_text_samples(dataset, batch_size_example)
    print(text_columns, text_content)
```
